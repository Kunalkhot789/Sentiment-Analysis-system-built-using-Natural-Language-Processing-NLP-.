```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
stop = set(stopwords.words('english'))
stop
```

```python
stem = PorterStemmer()
lemma = WordNetLemmatizer()
```

```python
def text_cleaner(text):
    text= str(text).lower()
    text= re.sub(r'\d+','',text)
    text= re.sub(r'[^\w\s]','',text)
    text=' '.join([word for word in text.split() if word not in stop])
    text=' '.join(stem.stem(word) for word in text.split())
    text=' '.join(lemma.lemmatize(word) for word in text.split())
    return text
```

```python
text_cleaner("Statistics is a easy and very useful, language and ! has easiest understanding")
```

```python
df["Cleaned Comments"] = df['Comment'].apply(text_cleaner)
df
```

```python
sentence = df['Cleaned Comments']
sentence
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
clean_sentences = [sentences for sentences in sentence]
clean_sentences
```

```python
tokenizer.fit_on_texts(clean_sentences)
word_index = tokenizer.word_index
word_index
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
sequences = tokenizer.texts_to_sequences(clean_sentences)
maxlen = max([len(seq) for seq in sequences])
padded_sequences = pad_sequences(sequences, maxlen=maxlen)
padded_sequences.shape
```

```python
padded_sequences[0]
```

```python
clean_sentences[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN,LSTM, Dense
embedding_dim= 16
vocab_size= len(word_index)+1
max_length= padded_sequences.shape[1]

model= Sequential([
    Embedding(input_dim= vocab_size,output_dim= embedding_dim,input_length= max_length),
                  SimpleRNN(32),
                  Dense(4,activation= 'softmax')])
```

```python
labels = df['Feedback']
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(padded_sequences, labels, epochs=10, verbose=1)
```

```python
import gradio as gr
def predict_sentiment(text):
    text= text_cleaner(text)
    test_sequences= tokenizer.texts_to_sequences([text])
    padded_test_sequences= pad_sequences(test_sequences, maxlen= maxlen)
    predicted_class= model.predict(padded_test_sequences)
    predicted_label= np.argmax(predicted_class)
    predicted_label= le.inverse_transform ([predicted_label])
    return predicted_label[0]
```

```python
predict_sentiment('Borderland')
```
